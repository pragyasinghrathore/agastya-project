```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings

EMBEDDING_MODEL = "BAAI/bge-large-en-v1.5"
COLLECTION_NAME = "itihasa_sanskrit"
CHROMA_DIR = "/content/drive/MyDrive/Agastya_Project/chromadb"

embedding_function = SentenceTransformerEmbeddings(
    model_name=EMBEDDING_MODEL
)

vectorstore = Chroma(
    collection_name=COLLECTION_NAME,
    embedding_function=embedding_function,
    persist_directory=CHROMA_DIR,
)

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",  # or another available chat model
    temperature=0.1
)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=(
        "Context from the Ramayana and related Itihāsa texts:\n"
        "{context}\n\n"
        "Question: {question}\n\n"
        "Answer in one or two short, factual sentences in English, "
        "using only the information in the context. "
        "If the context is insufficient, say that you cannot answer."
    )
)

def get_context(query: str) -> str:
    docs = retriever.invoke(query)
    chunks = []
    for d in docs:
        text = d.page_content
        if "\n\nEnglish:" in text:
            english = text.split("\n\nEnglish:")[-1].strip()
        else:
            english = text
        chunks.append(english)
    return "\n--\n".join(chunks)

def rag_generate_openai(query: str):
    context = get_context(query)
    prompt_text = prompt.format(context=context, question=query)
    response = llm.invoke(prompt_text)
    return response, context
```

```python
for q in [
    "Where does Agastya live?",
    "Who is wife of Rama?",
    "Who is son of Agastya?"
]:
    ans, ctx = rag_generate_openai(q)
    print("="*80)
    print("Q:", q)
    print("\nANSWER:\n", ans)
    print("\nCTX (truncated):\n", ctx[:400])
    print()
```

```python
def answer_accuracy(rag_fn, eval_items):
    total = len(eval_items)
    correct = 0
    for item in eval_items:
        q = item["query"]
        gold_keywords = [g.lower() for g in item["gold_keywords"]]
        allow_insufficient = item.get("allow_insufficient", False)

        pred, _ = rag_fn(q)
        pred_low = pred.content.lower()

        # Case 1: we expect a concrete answer
        if gold_keywords:
            if any(g in pred_low for g in gold_keywords):
                correct += 1
        # Case 2: we expect the model to say context is insufficient
        elif allow_insufficient:
            if "cannot answer" in pred_low or "insufficient" in pred_low or "not provide" in pred_low:
                correct += 1

    return correct / total if total > 0 else 0.0

print("OpenAI RAG answer accuracy:", answer_accuracy(rag_generate_openai, eval_answers))
```

```python
import math

# Example eval_queries – you already built this with real IDs
# Example – replace with your real IDs
eval_queries = [
    {
        "query": "Where does Sage Agastya live with Lopamudra?",
        "relevant_ids": [4658,26808]
    },
    {
        "query": "Who is the wife of Rama?",
        "relevant_ids": [5339,5852,8063]
    },
    {
        "query": "Who is the wife of Agastya?",
        "relevant_ids": [26146,26149,26139]
    },
    {
        "query": "Who is cursed by Agastya?",
        "relevant_ids": [596,27723,27725]
    },
    {
        "query": "Who is father of Agastya?",
        "relevant_ids": [70072,27723,27725]
    }
    # add 10–30 such entries
]

def _get_doc_id(doc):
    return (
        getattr(doc, "id", None)
        or doc.metadata.get("id")
        or doc.metadata.get("verse_id")
        or doc.metadata.get("source")
    )

def recall_at_k(retriever, eval_queries, k=5):
    scores = []
    for item in eval_queries:
        q = item["query"]
        relevant = set(item["relevant_ids"])
        docs = retriever.invoke(q)
        top_ids = [_get_doc_id(d) for d in docs[:k]]
        hits = len(relevant.intersection(top_ids))
        denom = len(relevant) if len(relevant) > 0 else 1
        scores.append(hits / denom)
    return sum(scores) / len(scores)

def mrr_at_k(retriever, eval_queries, k=5):
    scores = []
    for item in eval_queries:
        q = item["query"]
        relevant = set(item["relevant_ids"])
        docs = retriever.invoke(q)
        rr = 0.0
        for rank, d in enumerate(docs[:k], start=1):
            doc_id = _get_doc_id(d)
            if doc_id in relevant:
                rr = 1.0 / rank
                break
        scores.append(rr)
    return sum(scores) / len(scores)

def ndcg_at_k(retriever, eval_queries, k=5):
    scores = []
    for item in eval_queries:
        q = item["query"]
        relevant = set(item["relevant_ids"])
        docs = retriever.invoke(q)

        gains = []
        for d in docs[:k]:
            doc_id = _get_doc_id(d)
            gains.append(1.0 if doc_id in relevant else 0.0)

        dcg = sum(g / math.log2(i + 2) for i, g in enumerate(gains))
        ideal_gains = sorted(gains, reverse=True)
        idcg = sum(g / math.log2(i + 2) for i, g in enumerate(ideal_gains)) or 1.0
        scores.append(dcg / idcg)
    return sum(scores) / len(scores)

k = 5
print("Recall@{}:".format(k), recall_at_k(retriever, eval_queries, k))
print("MRR@{}:".format(k), mrr_at_k(retriever, eval_queries, k))
print("nDCG@{}:".format(k), ndcg_at_k(retriever, eval_queries, k))
```
